```python
# Program that receives age and experience from staff and returns their specified Annual Tax Revenue (ATR)

def calculate_atr(experience, age):
    if experience > 25 and age >= 55:
        return 5_600_000
    elif experience > 20 and age >= 45:
        return 4_480_000
    elif experience > 10 and age >= 35:
        return 1_500_000
    else:
        return 550_000

def main():
    print("Izifin Technology - Staff Annual Tax Revenue (ATR) Calculator")
    print("-----------------------------------------------------------")
    
    try:
        years_exp = float(input("Enter staff's years of work experience: "))
        age = float(input("Enter staff's age: "))
        
        if years_exp < 0 or age < 0:
            print("Error: Years of experience and age must be non-negative.")
            return
        
        atr = calculate_atr(years_exp, age)
        print(f"\nAnnual Tax Revenue (ATR): N{atr:,}")
        
    except ValueError:
        print("Error: Please enter valid numeric values.")

main()
```
