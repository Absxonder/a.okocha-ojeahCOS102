```python
#Flowchart design for simple interest calculator


'''

                                        Start
                                          |  
                                          |                                          
                                    Input P, R, T
                                          |  
                                          |   
                               Calculate A = P*(1 + (R*T)/100)
                                          |  
                                          |  
                                     Display (A)  
                                          |  
                                          |                                          
                                         End   


'''



#Input values

P = float(input("Enter the principal amount: "))
R = float(input("Enter the rate (%): "))
T = float(input("Enter the time in years: "))

#Calculate amount
A = (P*(1 + (R*T)/100))    

#Display result
print("Final amount = ", A)

            
```

```python
#Flowchart design for Compound interest calculator
'''

                                        Start
                                          |  
                                          |                                          
                                    Input P, R, n, T
                                          |  
                                          |   
                              Calculate A = P(1 + R/n)^(nt)
                                          |  
                                          |  
                                     Display (A)  
                                          |  
                                          |                                          
                                         End   


'''

#Input values

P = float(input("Enter principal amount: "))
R = float(input("Enter the interest rate %: "))
n = float(input("Enter the number of times compunded per year: "))
T = float(input("Enter the time in years: "))

#Calculate Amount

A = P * (1 + R/n) ** (n*T)

#Display result
print("Compounded Amount = ", A)
                      
```

```python
#Flowchart design for Annuity calculator


'''

                                        Start
                                          |  
                                          |                                          
                                    Input P, R, T
                                          |  
                                          |   
                    Calculate A = PMT * ((1 + R/n)^(nt) - 1) / (R/n)
                                          |  
                                          |  
                                      Display (A)  
                                          |  
                                          |                                          
                                         End   


'''
#Input values

PMT = float(input("Enter payment per period: "))
R = float(input("Enter interest rate %: "))
n = float(input("Enter number of times per year: "))
t = float(input("Enter time in years: "))

#Calcuate Amount

A = PMT * ((1 + R/n) ** (n*t) - 1) / (R/n)

#Display result

print("Future value = ", A)


```
